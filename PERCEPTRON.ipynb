```python
import numpy as np

class Perceptron:
    def __init__(self, input_size, learning_rate=0.1, epochs=10):
        self.weights = np.zeros(input_size)
        self.bias = 0
        self.lr = learning_rate
        self.epochs = epochs

    def activation(self, x):
        return 1 if x >= 0 else 0

    def predict(self, x):
        linear_output = np.dot(self.weights, x) + self.bias
        return self.activation(linear_output)

    def train(self, X, y):
        for epoch in range(self.epochs):
            print(f"Epoch {epoch+1}")
            for xi, target in zip(X, y):
                prediction = self.predict(xi)
                error = target - prediction
                # Update rule
                self.weights += self.lr * error * xi
                self.bias += self.lr * error
                print(f"Input: {xi}, Prediction: {prediction}, Error: {error}, Weights: {self.weights}, Bias: {self.bias}")
            print("")
    def accuracy(self, X, y):
        predictions = [self.predict(xi) for xi in X]
        correct = sum(p == t for p, t in zip(predictions, y))
        return correct / len(y)


# Sample Dataset: AND logic
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 0, 0, 1])  # AND logic output

# Create and train perceptron
perceptron = Perceptron(input_size=2)
perceptron.train(X, y)


```

```python
acc = perceptron.accuracy(X, y)
print(f"Training Accuracy: {acc * 100:.2f}%")
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from IPython.display import clear_output
import time

# Step 1: Generate training data
def generate_training_data():
    levels = np.arange(0, 121, 1)
    actions = []
    for level in levels:
        if level < 20:
            actions.append(20)
        elif level >= 100:
            actions.append(-1)
        else:
            actions.append(1)
    return pd.DataFrame({'level': levels, 'action': actions})

df = generate_training_data()

# Step 2: Preprocess
X = df[['level']].values / 100.0
y = df['action'].values

# Step 3: Build and train model
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(16, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1)
])
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
model.fit(X, y, epochs=100, batch_size=8, verbose=0)
model.save("smart_water_tank_model.h5")
print("âœ… Model trained and saved to smart_water_tank_model.h5")

# Step 4: SmartWaterTank Class
class SmartWaterTank:
    def __init__(self, model, min_level=0, max_level=100):
        self.level = 0
        self.min_level = min_level
        self.max_level = max_level
        self.model = model
        self.increasing = True  # starts by filling
        self.level_history = []

    def get_action(self):
        input_level = np.array([[self.level / 100.0]])
        predicted = self.model.predict(input_level, verbose=0)[0][0]
        if predicted < 0:
            return -1
        elif predicted < 10:
            return 1
        else:
            return 20

    def update_level(self):
        action = self.get_action()

        # Rule-based override to enforce threshold behavior
        if not self.increasing and self.level <= 20:
            self.increasing = True
        if self.level >= self.max_level:
            self.increasing = False

        # Adjust action to match direction
        if self.increasing:
            if action == -1:
                action = 1
        else:
            if action in [1, 20]:
                action = -1

        self.level += action
        self.level = max(self.min_level, min(self.level, self.max_level))
        self.level_history.append(self.level)
        return action

    def simulate(self, steps=400, delay=0.05):
        for step in range(steps):
            action = self.update_level()
            clear_output(wait=True)

            # Plotting
            plt.figure(figsize=(10, 5))
            plt.plot(self.level_history, color='blue', marker='o')
            plt.title("ðŸ“ˆ Smart Water Tank Level Over Time")
            plt.xlabel("Step")
            plt.ylabel("Water Level (mL)")
            plt.ylim(0, self.max_level + 10)
            plt.grid(True)
            plt.text(len(self.level_history)-1, self.level_history[-1],
                     f'{self.level_history[-1]:.1f} mL', fontsize=12,
                     verticalalignment='bottom')
            plt.show()

            print(f"Step {step+1}: Water Level = {self.level:.1f} mL \t(Action: {action:+})")
            time.sleep(delay)


# Step 5: Run simulation
tank = SmartWaterTank(model)
tank.simulate(steps=400, delay=0.2)
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from collections import deque

# Step 1: Generate enhanced training data with sequences
def generate_lstm_training_data():
    """Generate training sequences that include temporal patterns"""
    np.random.seed(42)

    # Simulate realistic water tank usage over time
    time_steps = 2000
    levels = []
    actions = []
    usage_patterns = []

    current_level = 50  # Start at middle level

    for t in range(time_steps):
        # Simulate daily usage pattern (higher usage during day)
        hour_of_day = (t % 24)
        usage_multiplier = 1.5 if 6 <= hour_of_day <= 22 else 0.5

        # Add some randomness
        noise = np.random.normal(0, 2)

        # Determine optimal action based on level and usage pattern
        if current_level < 15:
            action = 25  # Emergency fill
        elif current_level < 30:
            action = 10  # Fast fill
        elif current_level > 85:
            action = -5  # Drain
        elif current_level > 70:
            action = -2  # Slow drain
        else:
            # Predictive filling based on expected usage
            if usage_multiplier > 1.0:  # High usage period
                action = 5
            else:
                action = 1

        levels.append(current_level)
        actions.append(action)
        usage_patterns.append(usage_multiplier)

        # Update level (simulate consumption + action)
        consumption = usage_multiplier * 2 + noise
        current_level = max(0, min(100, current_level + action - consumption))

    return np.array(levels), np.array(actions), np.array(usage_patterns)

# Step 2: Create sequences for LSTM training
def create_lstm_sequences(levels, actions, usage_patterns, sequence_length=10):
    """Create input sequences and corresponding targets for LSTM"""
    X, y = [], []

    for i in range(sequence_length, len(levels)):
        # Input: sequence of [level, usage_pattern, previous_action]
        sequence = []
        for j in range(i-sequence_length, i):
            sequence.append([
                levels[j] / 100.0,  # Normalize level
                usage_patterns[j],   # Usage pattern
                actions[j-1] / 25.0 if j > 0 else 0  # Previous action (normalized)
            ])

        X.append(sequence)
        y.append(actions[i] / 25.0)  # Normalize target action

    return np.array(X), np.array(y)

# Generate training data
print("ðŸ”„ Generating LSTM training data...")
levels, actions, usage_patterns = generate_lstm_training_data()
X, y = create_lstm_sequences(levels, actions, usage_patterns, sequence_length=15)

print(f"âœ… Generated {X.shape[0]} training sequences")
print(f"ðŸ“Š Input shape: {X.shape}, Output shape: {y.shape}")

# Step 3: Build LSTM model
def build_lstm_model(sequence_length, features):
    """Build LSTM model for water tank control"""
    model = tf.keras.models.Sequential([
        # LSTM layers for temporal pattern learning
        tf.keras.layers.LSTM(64, return_sequences=True, input_shape=(sequence_length, features)),
        tf.keras.layers.Dropout(0.2),

        tf.keras.layers.LSTM(32, return_sequences=False),
        tf.keras.layers.Dropout(0.2),

        # Dense layers for final decision making
        tf.keras.layers.Dense(32, activation='relu'),
        tf.keras.layers.Dense(16, activation='relu'),
        tf.keras.layers.Dense(1)  # Single output for action
    ])

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        loss='mse',
        metrics=['mae']
    )

    return model

# Build and train model
print("ðŸ—ï¸ Building LSTM model...")
model = build_lstm_model(sequence_length=15, features=3)

print("ðŸ“ˆ Training LSTM model...")
history = model.fit(
    X, y,
    epochs=50,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)

# Save model
model.save("lstm_water_tank_model.h5")
print("âœ… LSTM model trained and saved!")

# Step 4: Enhanced Smart Water Tank with LSTM
class LSTMSmartWaterTank:
    def __init__(self, model, sequence_length=15, min_level=0, max_level=100):
        self.level = 50  # Start at middle level
        self.min_level = min_level
        self.max_level = max_level
        self.model = model
        self.sequence_length = sequence_length

        # History tracking
        self.level_history = deque(maxlen=1000)
        self.action_history = deque(maxlen=1000)
        self.usage_history = deque(maxlen=1000)

        # Initialize with some default values
        for _ in range(sequence_length):
            self.level_history.append(self.level / 100.0)
            self.action_history.append(0)
            self.usage_history.append(1.0)

    def get_current_usage_pattern(self, step):
        """Simulate realistic usage patterns"""
        hour_of_day = (step % 24)
        # Higher usage during day (6 AM to 10 PM)
        if 6 <= hour_of_day <= 22:
            base_usage = 1.5
        else:
            base_usage = 0.5

        # Add some weekly pattern (higher on weekends)
        day_of_week = (step % 168) // 24
        if day_of_week in [5, 6]:  # Weekend
            base_usage *= 1.2

        # Add some randomness
        return base_usage + np.random.normal(0, 0.1)

    def get_lstm_action(self, current_usage):
        """Get action prediction from LSTM model"""
        # Prepare sequence for LSTM input
        sequence = []
        recent_levels = list(self.level_history)[-self.sequence_length:]
        recent_usage = list(self.usage_history)[-self.sequence_length:]
        recent_actions = list(self.action_history)[-self.sequence_length:]

        for i in range(self.sequence_length):
            sequence.append([
                recent_levels[i],
                recent_usage[i],
                recent_actions[i] / 25.0
            ])

        # Get LSTM prediction
        input_seq = np.array([sequence])
        predicted_action = self.model.predict(input_seq, verbose=0)[0][0]

        # Denormalize and convert to discrete action
        action = predicted_action * 25.0

        # Convert to discrete actions
        if action < -10:
            return -15
        elif action < -2:
            return -5
        elif action < 2:
            return 0
        elif action < 15:
            return 10
        else:
            return 25

    def apply_safety_rules(self, action):
        """Apply safety rules to prevent overflow/underflow"""
        # Emergency overrides
        if self.level < 10:
            return max(15, action)  # Force significant filling
        elif self.level > 90:
            return min(-5, action)  # Force draining
        elif self.level < 25 and action < 0:
            return 5  # Prevent draining when low
        elif self.level > 75 and action > 10:
            return 0  # Prevent overfilling when high

        return action

    def update_level(self, step):
        """Update water level based on LSTM prediction and usage"""
        # Get current usage pattern
        current_usage = self.get_current_usage_pattern(step)

        # Get LSTM action recommendation
        lstm_action = self.get_lstm_action(current_usage)

        # Apply safety rules
        final_action = self.apply_safety_rules(lstm_action)

        # Simulate water consumption
        consumption = current_usage * 2 + np.random.normal(0, 0.5)

        # Update level
        self.level += final_action - consumption
        self.level = max(self.min_level, min(self.level, self.max_level))

        # Update histories
        self.level_history.append(self.level / 100.0)
        self.action_history.append(final_action / 25.0)
        self.usage_history.append(current_usage)

        return final_action, consumption, current_usage

    def simulate(self, steps=500):
        """Run simulation with real-time plotting"""
        plt.ion()
        fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 10))

        # Initialize plots
        levels_plot = []
        actions_plot = []
        usage_plot = []

        for step in range(steps):
            action, consumption, usage = self.update_level(step)

            levels_plot.append(self.level)
            actions_plot.append(action)
            usage_plot.append(usage)

            print(f"Step {step+1:3d}: Level={self.level:5.1f}mL | Action={action:+3.0f} | Usage={usage:.2f} | Consumption={consumption:.1f}")

            # Update plots every 10 steps
            if step % 10 == 0:
                # Clear and update plots
                ax1.clear()
                ax1.plot(levels_plot, 'b-', linewidth=2, label='Water Level')
                ax1.axhline(y=20, color='r', linestyle='--', alpha=0.7, label='Low Threshold')
                ax1.axhline(y=80, color='g', linestyle='--', alpha=0.7, label='High Threshold')
                ax1.set_ylabel('Water Level (mL)')
                ax1.set_title('LSTM-Controlled Smart Water Tank')
                ax1.legend()
                ax1.grid(True, alpha=0.3)

                ax2.clear()
                ax2.plot(actions_plot, 'r-', linewidth=1.5, label='Actions')
                ax2.axhline(y=0, color='k', linestyle='-', alpha=0.3)
                ax2.set_ylabel('Action (mL)')
                ax2.legend()
                ax2.grid(True, alpha=0.3)

                ax3.clear()
                ax3.plot(usage_plot, 'g-', linewidth=1.5, label='Usage Pattern')
                ax3.set_ylabel('Usage Multiplier')
                ax3.set_xlabel('Time Steps')
                ax3.legend()
                ax3.grid(True, alpha=0.3)

                plt.tight_layout()
                fig.canvas.draw()
                fig.canvas.flush_events()
                plt.pause(0.01)

        plt.ioff()
        plt.show()

        # Print performance summary
        optimal_range = sum(1 for level in levels_plot if 30 <= level <= 70)
        print(f"\nðŸ“Š Performance Summary:")
        print(f"   Time in optimal range (30-70mL): {optimal_range/len(levels_plot)*100:.1f}%")
        print(f"   Average level: {np.mean(levels_plot):.1f}mL")
        print(f"   Level stability (std): {np.std(levels_plot):.1f}mL")

# Step 5: Run LSTM-based simulation
print("\nðŸš€ Starting LSTM Water Tank Simulation...")
lstm_tank = LSTMSmartWaterTank(model)
lstm_tank.simulate(steps=300)
```
