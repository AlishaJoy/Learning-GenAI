```python
from nltk.util import ngrams
```

```python
sample_text = 'I am learning NLP(Natural Language Processing)'
tokens = word_tokenize(sample_text)
```

```python
# Unigram
unigrams = list(ngrams(tokens, 1))
print('Unigrams:', unigrams)

# Bigram
bigrams = list(ngrams(tokens, 2))
print('Bigrams:', bigrams)

# Trigram
trigrams = list(ngrams(tokens, 3))
print('Trigrams:', trigrams)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
texts = [
  'I love programming.', 'Python is amazing.',
  'I enjoy machine learning.', 'The weather is nice today.', 'I like algo.',
  'Machine learning is fascinating.', 'Natural Language Processing is a part of AI.'
]

labels = [
  'tech', 'tech', 'tech', 'non-tech', 'tech', 'tech', 'tech'
]
```

```python
vectorizer = CountVectorizer()
x = vectorizer.fit_transform(texts)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, labels, test_size=0.2, random_state=42)
```

```python
model = MultinomialNB()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
reviews = ['This movie was fantastic! A must-watch.',
           'I didn\'t enjoy this movie at all.',
           'Amazing storyline and great acting!',
           'The plot was dull and predictable.',
           'Loved the cinematography! Highly recommended.']

labels = ['positive', 'negative', 'positive', 'negative', 'positive']
```

```python
vectorizer = CountVectorizer()
x = vectorizer.fit_transform(reviews)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, labels, test_size=0.2, random_state=42)
```

```python
model = MultinomialNB()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)

print('Accuracy:', accuracy)
```

```python
if accuracy > 0.8:
  print('Good vibes. Book the ticket!')
else:
  print('Needs more work!')
```
