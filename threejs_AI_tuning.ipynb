```python
from langchain.document_loaders import DirectoryLoader, BSHTMLLoader

# Load all HTML files from "threejs_docs" folder
loader = DirectoryLoader("https://threejs.org/", glob="*.html", loader_cls=BSHTMLLoader)
documents = loader.load()

print(f"Loaded {len(documents)} documents")
```

```python
import os
from langchain_community.document_loaders import BeautifulSoupLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Step 1: Prepare Your Three.js Documents (The Knowledge Base)

# Gather Data: Specify the directory containing your Three.js documentation
docs_directory = "/content/threejs_docs" # Replace with the actual path to your documents

# Ensure the directory exists (create it if it doesn't and add your files)
os.makedirs(docs_directory, exist_ok=True)

# Load Documents: Use BeautifulSoupLoader to load HTML documents
# You can extend this to handle other document types as needed (e.g., PDF, text)
documents = []
for file_name in os.listdir(docs_directory):
    if file_name.endswith(".html"):
        file_path = os.path.join(docs_directory, file_name)
        loader = BeautifulSoupLoader(file_path)
        documents.extend(loader.load())

# Chunk Data: Split the large documents into smaller, manageable text chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)
chunked_documents = text_splitter.split_documents(documents)

print(f"Loaded {len(documents)} documents and split into {len(chunked_documents)} chunks.")
# You can inspect the first few chunks to see the result
# print(chunked_documents[:5])
```
